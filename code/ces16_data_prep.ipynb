## Recode numeric variables into categorical variables that match the 2020 and 2024 datasets

```python
ces16_df = ces16_subset2.copy().drop(columns=['tookpost'])
# recode numeric variables
# state
fips_to_state = {
    1: "Alabama",
    2: "Alaska",
    4: "Arizona",
    5: "Arkansas",
    6: "California",
    8: "Colorado",
    9: "Connecticut",
    10: "Delaware",
    11: "District of Columbia",
    12: "Florida",
    13: "Georgia",
    15: "Hawaii",
    16: "Idaho",
    17: "Illinois",
    18: "Indiana",
    19: "Iowa",
    20: "Kansas",
    21: "Kentucky",
    22: "Louisiana",
    23: "Maine",
    24: "Maryland",
    25: "Massachusetts",
    26: "Michigan",
    27: "Minnesota",
    28: "Mississippi",
    29: "Missouri",
    30: "Montana",
    31: "Nebraska",
    32: "Nevada",
    33: "New Hampshire",
    34: "New Jersey",
    35: "New Mexico",
    36: "New York",
    37: "North Carolina",
    38: "North Dakota",
    39: "Ohio",
    40: "Oklahoma",
    41: "Oregon",
    42: "Pennsylvania",
    44: "Rhode Island",
    45: "South Carolina",
    46: "South Dakota",
    47: "Tennessee",
    48: "Texas",
    49: "Utah",
    50: "Vermont",
    51: "Virginia",
    53: "Washington",
    54: "West Virginia",
    55: "Wisconsin",
    56: "Wyoming"
}

# gender
gender_map_numeric = {
    1: "Male",
    2: "Female"
}

# race
race_map_numeric = {
    1: 'White',
    2: 'Black',
    3: 'Hispanic',
    4: 'All Other Races',
    5: 'All Other Races',
    6: 'All Other Races',
    7: 'All Other Races',
    8: 'All Other Races'
}

# economy
econ_map_numeric = {
    1: 'Gotten much better',
    2: 'Gotten better',
    3: 'Stayed the same',
    4: 'Gotten worse',
    5: 'Gotten much worse',
    6: 'Not sure'
}

# income
income_map_numeric = {
    1: 'Increased a lot',
    2: 'Increaseed somewhat',
    3: 'Stayed the same',
    4: 'Decreased somewhat',
    5: 'Decreased a lot'
}

# obama approval
obama_map_numeric = {
    1: 'Strongly approve',
    2: 'Somewhat approve',
    3: 'Somewhat disapprove',
    4: 'Strongly disapprove',
    5: 'Not sure'
}

educ_map_numeric = {
    1: 'No HS',
    2: 'High school graduate',
    3: 'Some college',
    4: '2-year',
    5: '4-year',
    6: 'Post-grad'
}

# recode numeric variables
ces16_df['inputstate'] = ces16_df['inputstate'].map(fips_to_state)
ces16_df['gender'] = ces16_df['gender'].map(gender_map_numeric)
ces16_df['race'] = ces16_df['race'].map(race_map_numeric)
ces16_df['CC16_302'] = ces16_df['CC16_302'].map(econ_map_numeric)
ces16_df['CC16_303'] = ces16_df['CC16_303'].map(income_map_numeric)
ces16_df['CC16_320a'] = ces16_df['CC16_320a'].map(obama_map_numeric)
ces16_df['educ'] = ces16_df['educ'].map(educ_map_numeric)

# create an age variable
ces16_df['age'] = 2016 - ces16_df['birthyr']

# create a vote_clinton variable
ces16_df['presvote16'] = np.where(ces16_df['CC16_410a'] == 1.0, 'Donald Trump (Republican)', 'Hillary Clinton (Democrat)')
ces16_df['vote_clinton'] = np.where(ces16_df['CC16_410a'] == 2.0, 1, 0)
ces16_df['vote_trump'] = np.where(ces16_df['CC16_410a'] == 1.0, 1, 0)
```

```python
# rename columns
ces16_df = ces16_df.rename(columns={'inputstate': 'state',
                         'CC16_302': 'econ_past_year',
                         'CC16_303': 'income_past_4yrs',
                         'CC16_320a': 'incumbent_pres_approval',
                         })
```

```python
# make columns categorical
gender_categories = pd.CategoricalDtype(categories=['Male', 'Female'], ordered=True)
age_categories = pd.CategoricalDtype(categories=['18–24', '25–34', '35–44', '45–54', '55–64', '65–74', '75+'], ordered=True)

ces16_df['presvote16'] = ces16_df['presvote16'].astype('category')
ces16_df['age_bracket'] = ces16_df['age_bracket'].astype(age_categories)
ces16_df['gender'] = ces16_df['gender'].astype(gender_categories)
ces16_df['race'] = ces16_df['race'].astype('category')
# ces16_df['state'] = ces16_df['state'].astype('category')
ces16_df['region'] = ces16_df['region'].astype('category')
ces16_df['econ_past_year'] = ces16_df['econ_past_year'].astype('category')
ces16_df['income_past_4yrs'] = ces16_df['income_past_4yrs'].astype('category')
ces16_df['incumbent_pres_approval'] = ces16_df['incumbent_pres_approval'].astype('category')
```

```python
ces16_clean = ces16_df[['presvote16',
                        'vote_clinton',
                        'vote_trump',
                        'age_bracket',
                        'race',
                        'gender',
                        'educ',
                        'state',
                        'region',
                        'econ_past_year',
                        'income_past_4yrs',
                        'incumbent_pres_approval'
                        ]]
```

```python
# create dummies
ces16_clean2 = ces16_clean.copy().drop(columns=['presvote16'])

def create_dummy_interaction(col1, col2):
  result = pd.DataFrame()
  col1_dummies = pd.get_dummies(col1, drop_first=True)
  col2_dummies = pd.get_dummies(col2, drop_first=True)
  for col1_val in col1_dummies.columns:
    for col2_val in col2_dummies.columns:
      result[f'{col1_val}_{col2_val}'] = col1_dummies[col1_val] * col2_dummies[col2_val]
  return result

# create the interaction dummies
race_gender_interaction_terms = create_dummy_interaction(ces16_clean2['race'], ces16_clean2['gender'])
race_educ_interaction_terms = create_dummy_interaction(ces16_clean2['race'], ces16_clean2['educ'])
race_region_interaction_terms = create_dummy_interaction(ces16_clean2['race'], ces16_clean2['region'])

ces16_dummies = pd.concat([
    pd.get_dummies(ces16_clean2, drop_first=True),
    race_gender_interaction_terms,
    race_educ_interaction_terms,
    race_region_interaction_terms
], axis=1)
```
